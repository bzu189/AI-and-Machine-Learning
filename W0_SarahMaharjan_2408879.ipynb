# **Task 1 – Unit Conversion Program**
**Question:**

Create a Python program that converts between different units of measurement.

* The program should:

  1. Prompt the user to choose the type of conversion.
  2. Ask the user to input the value to be converted.
  3. Perform the conversion and display the result.
  4. Handle potential errors such as invalid input.
  5. Use functions and include proper docstrings.

* Requirements:
  1. Functions: Define at least one function to perform the conversion.
  2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
  3. User Input: Prompt the user to select the conversion type and input the value.
  4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
  return value
* Conversion Options:
    1. Length:

      – Convert meters (m) to feet (ft).

      – Convert feet (ft) to meters (m).

    2. Weight:

      – Convert kilograms (kg) to pounds (lbs).

      – Convert pounds (lbs) to kilograms (kg).

    3. Volume:

      – Convert liters (L) to gallons (gal).

      – Convert gallons (gal) to liters (L).

```python
def unit_converter(conversion_type, value):
    """
    Converts units between length, weight, and volume.

    Parameters:
    conversion_type (str): Type of conversion
    value (float): Value to convert

    Returns:
    float: Converted value
    """
    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "L_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_L":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type.")

try:
    print("Available conversions:")
    print("m_to_ft, ft_to_m, kg_to_lbs, lbs_to_kg, L_to_gal, gal_to_L")

    choice = input("Enter conversion type: ")
    value = float(input("Enter value: "))

    result = unit_converter(choice, value)
    print("Converted value:", result)

except ValueError as e:
    print("Error:", e)
```

# Task 2 – Mathematical Operations on a List
Question:

**Create a program that:**

1. Prompts the user to choose an operation (sum, average, maximum, minimum).

2. Accepts a list of numbers.

3. Performs the selected operation.

4. Handles invalid input using try-except.

5. Uses separate functions with docstrings.

```python
def calculate_sum(numbers):
    """Returns the sum of a list of numbers."""
    return sum(numbers)

def calculate_average(numbers):
    """Returns the average of a list of numbers."""
    return sum(numbers) / len(numbers)

def calculate_max(numbers):
    """Returns the maximum number."""
    return max(numbers)

def calculate_min(numbers):
    """Returns the minimum number."""
    return min(numbers)

try:
    operation = input("Choose operation (sum, avg, max, min): ")
    numbers = list(map(float, input("Enter numbers separated by space: ").split()))

    if not numbers:
        raise ValueError("List cannot be empty")

    if operation == "sum":
        print("Result:", calculate_sum(numbers))
    elif operation == "avg":
        print("Result:", calculate_average(numbers))
    elif operation == "max":
        print("Result:", calculate_max(numbers))
    elif operation == "min":
        print("Result:", calculate_min(numbers))
    else:
        print("Invalid operation")

except ValueError as e:
    print("Error:", e)
```

# Section 4.3 – Nested List Exercises

**Flatten Nested List**

```python
def flatten(lst):
    flat = []
    for item in lst:
        if isinstance(item, list):
            flat.extend(flatten(item))
        else:
            flat.append(item)
    return flat

print(flatten([[1,2],[3,[4,5]],6]))
```

**Access Nested Element**

```python
def access_nested_element(lst, indices):
    for index in indices:
        lst = lst[index]
    return lst

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))
```

**Sum Nested List**


```python
def sum_nested(lst):
    return sum(flatten(lst))

print(sum_nested([[1,2],[3,[4,5]],6]))
```

**Remove Element from Nested List**

```python
def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

**Find Max in Nested List**

```python
def find_max(lst):
    return max(flatten(lst))

print(find_max([[1,2],[3,[4,5]],6]))
```

**Count Occurrences**

```python
def count_occurrences(lst, elem):
    return flatten(lst).count(elem)

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

**Deep Flatten**

```python
def deep_flatten(lst):
    return flatten(lst)
```

**Average Nested**

```python
def average_nested(lst):
    flat = flatten(lst)
    return sum(flat)/len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```
